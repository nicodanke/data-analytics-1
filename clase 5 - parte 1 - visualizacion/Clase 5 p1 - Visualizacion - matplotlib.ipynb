Concepto de **Artist**: 

Todos los gráficos de matplotlib están compuestos por su figura y sus axes. Esto es lo primero que tenemos que crear 

- **Figura**: la figura, puede interpretarse como lienzo o un contenedor de todos los objetos del gráfico. La forma más simple de crear una figura es usando pyplot: 
   ``` python  
     fig = plt.figure() 
    ```
- **Artist**: básicamente, todo lo que es visible en la Figura es un Artist (incluso los objetos Figure, Axes y Axis). Cuando se renderiza la figura, todos los Artist se dibujan en el lienzo. 


- **Axes**: es un Artist adjunto a una Figura que contiene una región para trazar datos, y normalmente incluye dos (o tres en el caso de 3D) objetos Axis. Podemos crear un Axes del siguente modo:
   ``` python  
     ax = plt.axes() 
    ```
- **Axis**: Estos objetos establecen la escala y los límites y generan ticks (las marcas en el Axis) y ticklabels (cadenas que etiquetan los ticks).  


````python
fig = plt.figure()
ax = plt.axes()

ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);

plt.show()
````

````python
fig, ax = plt.subplots()  # Crear una Figura que contenga un solo Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Graficamos datos en el Axes

plt.show()
````

La función `plt.subplots()` nos servirá más adelante para crear múltiples gráficos en una figura. Por default, genera un solo gráfico por fila y por columna.

Podemos optar por dos estilos de programación. 

### Estilo Orientado a Objetos:

````python
x = np.linspace(0, 2, 100)  # Creamos los datos

fig, ax = plt.subplots(figsize=(9, 5)) # Con figsize regulamos el tamaño de la figura
ax.plot(x, x, label='linear')  # Graficamos datos en el axes
ax.plot(x, x**2, label='quadratic')  # Graficamos más datos en el axes
ax.plot(x, x**3, label='cubic')  # ... y más datos
ax.set_xlabel('x label')  # agregamos la etiqueta x
ax.set_ylabel('y label')  # agregamos la etiqueta y
ax.set_title("Simple Plot")  # Agregamos el título
ax.legend();  # y la leyenda

plt.show()
````

### Estilo pyplot

Delegar en pyplot la creación y gestión automática de las Figuras y los Axes, y utilizar las funciones de pyplot para graficar:

````python
x = np.linspace(0, 2, 100)  # Creamos los datos

plt.figure(figsize=(9, 5))
plt.plot(x, x, label='linear')  #  Graficamos los datos en el Axes implícito
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();

plt.show()
````

##### Color: 
Con el argumento `color` podemos definir el color de la línea. Podemos utilizar diferentes [formatos](https://matplotlib.org/stable/tutorials/colors/colors.html):
- RGB or RGBA (red, green, blue, alpha) tuplas de floats entre [0, 1], por ejemplo (0.1, 0.2, 0.5)
- Case-insensitive hex RGB or RGBA string: '#0f0f0f'
- Notación abreviada de un solo carácter para algunos colores básicos: 'b' (blue), 'g' (green), 'r' (red)

##### Estilo de linea:

Con el argumento `linestyle` podemos seleccionar el estilo de la linea. Por ejemplo:
- `linestyle = 'solid'` o `linestyle = '-'`  linea continua
- `linestyle = 'dashed'` o `linestyle = '--'` linea de trazos
- `linestyle = 'dashdot'` o `linestyle = '-.'` linea compuesta (punto y segmento)
- `linestyle = 'dotted'` o `linestyle = ':'` linea punteada

##### Podemos seleccionar el estilo y color simultáneamente:

- `'-b'` linea continua azul

- `'--g'` linea de trazon verde

- `'-.r'` linea compuesta roja

- `':k'` linea punteada negra


Por último con el argumento `linewidth` regulamos el grosor de la línea.


````python
plt.figure(figsize=(9, 5))
plt.plot(x, x, label='linear', color='blue', linestyle = 'solid', linewidth=1)  
plt.plot(x, x**2, label='quadratic', color=(1.0, 0.0, 0.0), linestyle = '--', linewidth=2) 
plt.plot(x, x**3, ':k', label='cubic', linewidth=3)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();

plt.show()
````

#### Scatter Plots

Un **scatter plot** o **gráfico de dispersión** es un tipo de diagrama matemático que utiliza las coordenadas cartesianas para mostrar los valores de dos variables para un conjunto de datos.

````python
np.random.seed(1)
data1, data2, data3, data4 = np.random.randn(4, 100) 
````

````python
fig, ax = plt.subplots(figsize=(9, 5))

ax.scatter(data1, data2, s=50, facecolor='C0', edgecolor='k', label='random scatter')
ax.legend(loc='best', fontsize=15)
ax.set_xlabel('x', fontsize=15) # Agregamos nombre a los ejes
ax.set_ylabel('f(x)', fontsize=15); # Agregamos nombre a los ejes

plt.show()
````

````python
fig, ax = plt.subplots(figsize=(9, 5))

ax.scatter(data1, data2, s=100, facecolor='g', edgecolor='k', label='random scatter', marker='d') # marker = d 
ax.legend(loc='lower left', fontsize=15)
ax.set_xlabel('x', fontsize=15) # Agregamos nombre a los ejes
ax.set_ylabel('f(x)', fontsize=15); # Agregamos nombre a los ejes
````

### Funciones de apoyo

Cuando vamos a realizar el mismo gráfico varias veces, resulta útil crear funciones.

````python
def my_plotter(ax, x, y, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.scatter(x, y, **param_dict)
    return out
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data1, data2, {'facecolor':'g', 'marker': 'o', 'alpha':0.5});
````

````python
def my_plotter(ax, x, y, title, param_dict):
    """
    A helper function to make a graph.
    """
    ax.scatter(x, y, **param_dict)
    ax.set_title(title)
    return None
````

````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(9, 5))

plt.subplots_adjust(wspace=0.4, 
                    hspace=0.4)

my_plotter(ax1, data1, data2, 'axis 1', {'marker': 'x', })
my_plotter(ax2, data1, data2, 'axis 2', {'facecolor':'g', 'marker': 'o', 'alpha':0.5})
my_plotter(ax3, data3, data4, 'axis 3', {'marker': 'x'})
my_plotter(ax4, data3, data4, 'axis 4', {'facecolor':'g', 'marker': 'o', 'alpha':0.5});
````

Notar que en el caso anterior generamos varios gráficos en una misma figura pasándole argumentos a la función `plt.subplots()`. 

Los argumentos que recibe son el número de filas y el número de columnas, y dos argumentos opcionales (sharex y sharey) que permite especificar la relación entre los ejes.

## Histogramas

Entre los parámetros que podemos establecer para un histograma están:

**bins** : entero, lista o string. 

* Si es un `entero` define la cantidad de bins (barras) de igual ancho en el rango de valores del gráfico. 

* Si es una `lista`, define los límites de cada barra incluyendo el límite izquierdo del primer bin y el límite derecho del último bin. En este caso, pueden ser de anchos distintos.

* Si es un `string`, representa una de las estrategias que soporta numpy.histogram_bin_edges: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', 'sqrt'.

**density**: booleano. Si es True, los pesos están normalizados tal que la integral de la densidad en el rango vale 1.

**histtype**: string. Tipo de histograma a graficar. Los valores posibles son {'bar', 'barstacked', 'step', 'stepfilled'}

* 'bar' representa un histograma tradicional

* 'barstacked' representa un histograma de barras apiladas

* 'step' genera un gráfico de lineas que por default no tiene relleno

* 'stepfilled' genera un gráfico de lineas que por default tiene relleno

**color**: color que toman las barras del histograma

**edgecolor**: color que toman los bordes de las barras del histograma

**alpha**: decimal de valor entre 0 y 1 que representa el nivel de transparecia, 0 es transparente y 1 es opaco


````python
np.random.seed(1)
data_normal = np.random.normal(loc=5, scale=1, size=1000);

type(data_normal)
````

````python


fig = plt.figure()
# creo los axes
ax = plt.axes()
# dibujo en axes:
ax.hist(data_normal, bins=25, label='data normal', alpha=0.5, density=True)
ax.legend();
````

Si vamos a establecer los mismos valores de un conjunto de los parámetros en todos los histogramas que componen el gráfico, es conveniente definir un diccionario con estos valores, y pasarlo como parámetro en todas las llamadas al métodos hist, en lugar de establecer cada uno de los valores de los argumentos para cada uno de los histogramas.

Vamos a establecer los valores de alpha, density, y bins en un diccionario y luego repetir el gráfico anterior.

````python
np.random.seed(1)

kwargs = {'alpha': 0.3, 
          'density': False, 
          'bins': 40}

data_normal_1 = np.random.normal(loc=5, scale=1.5, size=1000)
data_normal_2 = np.random.normal(loc=3, scale=1, size=1000)

fig = plt.figure()

# creo los axes
ax = plt.axes()

# dibujo en axes:
ax.hist(data_normal_1, label='data normal 1', **kwargs)

# superpongo otro gráfico en axes:
ax.hist(data_normal_2, label='data normal 2', **kwargs)
ax.legend();
````
