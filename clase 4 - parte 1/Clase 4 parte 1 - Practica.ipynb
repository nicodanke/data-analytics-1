### Ejercicio 6

* Generar una columna que contenga la hora de las llamadas

* Discretizar la hora de modo tal que se mapeen los rangos horarios a las siguientes categorías:

    - 0-5am - night
    - 5-12 - morning
    - 12-20 - afternoon
    - 20 - 0 evening
    
    
* Hacer una tabla pivot que muestre la duración media de las llamadas en función de la franja horaria y del mes

* Hacer una tabla pivot que muestre el número de llamadas en función de la franja horaria y del mes


```python
### Completa con tu código aquí
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y %H:%M')
df['hour'] = df['date'].dt.hour
df['hour_category'] = pd.cut(df['hour'], bins=[0, 5, 12, 20, 24], labels=['night', 'morning', 'afternoon', 'evening'])
df.head()
pv1 = df.groupby(['month', 'hour_category'])[['duration']].mean()
pv1.head(20)

pv2 = df.groupby(['month', 'hour_category'])[['duration']].count()
pv2.head(20)








```
